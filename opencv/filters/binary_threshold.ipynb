````python
def plot3_img(im1, im2, im3, t1,t2,t3):
    plt.figure(figsize=(10,10))
    plt.subplot(1,3,1)
    plt.imshow(im1)
    plt.title(t1)
    plt.subplot(1,3,2)
    plt.imshow(im2)
    plt.title(t2)
    plt.subplot(1,3,3)
    plt.imshow(im3)
    plt.title(t3)
````

````python
import cv2 as cv
imag = './image.png'
img = cv2.imread(imag)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
ret, binary = cv2.threshold(gray,128,maxval=255,type = cv2.THRESH_BINARY) # Returns the threshold used as well as the binary threshold image 
plt.imshow(binary,cmap ='gray') # cmap is gray
plt.show()
````

````python
plot3_img(img, gray,binary, "image","grayscale","binary threshold")
````

### Why is ret needed?

Normally, when you call:

`cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)`

you set the threshold manually (128).
So ret will just be 128.

But in automatic thresholding methods — like:

```
cv2.THRESH_OTSU
cv2.THRESH_TRIANGLE
```

OpenCV computes the best threshold value automatically,
and then ret will hold that calculated value.
