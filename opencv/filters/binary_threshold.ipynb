````python
import cv2
imag = '../image.png'
img = cv2.imread(imag)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(gray,150,maxval=255,type = cv2.THRESH_BINARY) # Returns the threshold used as well as the binary threshold image 
plt.imshow(binary) # cmap is gray
plt.show()
````

### Why is ret needed?

Normally, when you call:

`cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)`

you set the threshold manually (128).
So ret will just be 128.

But in automatic thresholding methods — like:

```
cv2.THRESH_OTSU
cv2.THRESH_TRIANGLE
```

OpenCV computes the best threshold value automatically,
and then ret will hold that calculated value.

````python

def plot4_img(a,b,c,d,p,q,r,s):
    plt.figure(figsize=(15,15))
    plt.subplot(2,2,1)
    plt.imshow(a)
    plt.title(p)
    plt.subplot(2,2,2)
    plt.imshow(b)
    plt.title(q)
    plt.subplot(2,2,3)
    plt.imshow(c)
    plt.title(r)
    plt.subplot(2,2,4)
    plt.imshow(d)
    plt.title(s)
    
    plt.show()

i = '../image.png'
img = cv2.imread(i)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gauss = cv2.GaussianBlur(gray, ksize=(51,51), sigmaX=1)
edges = cv2.Canny(gray, 150, 200)
plot4_img(img,gray,gauss,edges,"i","g","gau","ed")


````
